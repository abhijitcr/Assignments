Q12)  Below are the scores obtained by a student in tests 
34,36,36,38,38,39,39,40,40,41,41,41,41,42,42,45,49,56

1)	Find mean, median, variance, standard deviation.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the test scores
test_scores = [34, 36, 36, 38, 38, 39, 39, 40, 40, 41, 41, 41, 41, 42, 42, 45, 49, 56]
```

```python

# Calculate the mean and round to two decimal places
mean_score = round(np.mean(test_scores), 2)

# Calculate the median and round to two decimal places
median_score = round(np.median(test_scores), 2)

# Calculate the variance and round to two decimal places
variance_score = round(np.var(test_scores, ddof=1), 2)  # ddof=1 for sample variance

# Calculate the standard deviation and round to two decimal places
std_deviation = round(np.std(test_scores, ddof=1), 2)   # ddof=1 for sample standard deviation

# Print the rounded results
print("Mean Score:", mean_score)
print("Median Score:", median_score)
print("Variance:", variance_score)
print("Standard Deviation:", std_deviation)
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Create a histogram using Seaborn
sns.histplot(test_scores, bins=10, kde=True, color='skyblue')
plt.xlabel('Test Scores')
plt.ylabel('Frequency')
plt.title(f'Test Score Distribution\nMean: {mean_score}, Std Deviation: {std_deviation}')

# Show the plot
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(8, 5))
plt.boxplot(test_scores, vert=False)
plt.xlabel('Test Scores')
plt.title('Box Plot of Test Scores')
plt.grid(True)

# Show the plot
plt.show()
```
