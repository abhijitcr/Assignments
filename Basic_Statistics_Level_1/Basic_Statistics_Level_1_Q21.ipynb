Q 21) Check whether the data follows normal distribution

a)	Check whether the MPG of Cars follows Normal Distribution 

Dataset: Cars.csv

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cars = pd.read_csv('Cars.csv')
```

```python
cars.head()
```

```python
cars.tail()
```

```python
cars.info()
```

```python
cars.describe()
```

```python
# Create a histogram
plt.figure(figsize=(8, 4))
sns.histplot(cars['MPG'], kde=True, color='blue')
plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```

```python

# Create a Q-Q plot
plt.figure(figsize=(6, 6))
stats.probplot(cars['MPG'], dist="norm", plot=plt)
plt.title("Probability Plot (Q-Q plot) for MPG")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.show()
```

Answer (21_a): The MPG data is approximately normally distributed

Q21 b)	Check Whether the Adipose Tissue (AT) and Waist Circumference(Waist)  from wc-at data set  follows Normal Distribution

Dataset: wc-at.csv

```python
wcat = pd.read_csv('wc-at.csv')
```

```python
wcat.head()
```

```python
wcat.tail()
```

```python
wcat.info()
```

```python
wcat.describe()
```

```python
# Extract the 'AT' and 'Waist' columns
at_data = wcat['AT']
waist_data = wcat['Waist']

# Set up a 1x2 grid for subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot histograms for 'AT' and 'Waist'
sns.histplot(at_data, kde=True, color='blue', ax=axes[0])
axes[0].set_title("Histogram of AT")
axes[0].set_xlabel("AT")
axes[0].set_ylabel("Frequency")

sns.histplot(waist_data, kde=True, color='green', ax=axes[1])
axes[1].set_title("Histogram of Waist")
axes[1].set_xlabel("Waist")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Create Q-Q plots for 'AT' and 'Waist'
plt.figure(figsize=(12, 5))

# Q-Q plot for 'AT'
plt.subplot(1, 2, 1)
stats.probplot(at_data, dist="norm", plot=plt)
plt.title("Q-Q Plot for AT")

# Q-Q plot for 'Waist'
plt.subplot(1, 2, 2)
stats.probplot(waist_data, dist="norm", plot=plt)
plt.title("Q-Q Plot for Waist")

plt.tight_layout()
plt.show()
```

Answer (21_b): AT and Waist does not follows Normal Distribution
